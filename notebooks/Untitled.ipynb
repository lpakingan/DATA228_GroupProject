```python
import pandas as pd

# Load your results
df_lof = pd.read_parquet("../outputs/user_anomaly_lof_k100_stability.parquet")
# Corrected typo below: read_parquet
df_if = pd.read_parquet("../outputs/user_anomaly_scores_v1_c001_mf07.parquet")

def get_anomaly_profile(df, model_name):
    # Only look at users the model flagged as anomalies
    anomalies = df[df['is_anomaly'] == 1]
    
    profile = {
        "Model": model_name,
        "Total Anomalies": len(anomalies),
        "Avg Review Count": anomalies['review_count'].mean(),
        "Max Review Count": anomalies['review_count'].max(),
        "Avg Reviews/Day": anomalies['reviews_per_day'].mean(),
        "Avg Friends": anomalies['num_friends'].mean(),
        "% with 1 Review": (anomalies['review_count'] == 1).mean() * 100
    }
    return pd.Series(profile)

# Compare them side-by-side
comparison = pd.DataFrame([
    get_anomaly_profile(df_lof, "Local Outlier Factor (LOF)"),
    get_anomaly_profile(df_if, "Isolation Forest (IF)")
])

# Transpose for easier reading
print(comparison.set_index("Model").T)
```
